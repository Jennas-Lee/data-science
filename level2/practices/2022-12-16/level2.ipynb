```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import KMeans
```

```python
df_sub = df.iloc[:, 1:]

scaler = StandardScaler()
scaler.fit(df_sub)
df_std = pd.DataFrame(scaler.transform(df_sub), columns=df_sub.columns)
df_std
```

```python
df_sub = df.iloc[:, 1:]

scaler = StandardScaler()
scaler.fit(df_sub)
df_std = pd.DataFrame(scaler.transform(df_sub), columns=df_sub.columns)

for i in range(2, 6):
    df_kmeans = df_std
    model = KMeans(n_clusters=i, random_state=1234)
    model.fit(df_std)
    df_kmeans['cluster'] = model.labels_
    score = silhouette_score(X=df_kmeans.iloc[:, :-1], labels=df_kmeans.loc[:, ['cluster']], random_state=1234)
    print(i, score)

df_kmeans = df_std
model = KMeans(n_clusters=2, random_state=1234)
model.fit(df_std)
df_sub['cluster'] = model.labels_
df_sub.groupby('cluster')['ONEOFF_PURCHASES'].mean().max().round(2)
```

```python
df_train = df.loc[df.CUST_ID % 4 != 0, :]
df_test = df.loc[df.CUST_ID % 4 == 0, :]
print(df_train.shape)
print(df_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
col_list = df_train.columns.to_list()
col_list.remove('CUST_ID')
col_list.remove('ONEOFF_PURCHASES')
col_list
```

```python
model = DecisionTreeRegressor(max_depth=5, random_state=1234)
model.fit(X=df_train.loc[:, col_list], y=df_train.loc[:, ['ONEOFF_PURCHASES']])
pred = model.predict(X=df_test.loc[:, col_list])

df_measure = pd.DataFrame({'exact': df_test.loc[:, 'ONEOFF_PURCHASES'], 'pred': pred})
((((df_measure.exact - df_measure.pred) ** 2).sum() / df_measure.shape[0]) ** 0.5).round(0)
```

```python
weather_dataset = pd.read_csv('./../../../datasets/level3/weather.csv')
```

```python
weather_dataset.head()
```

```python
weather_dataset = weather_dataset.sort_values(by=['region_id', 'datetime'])
```

```python
weather_dataset.head()
```

```python
weather_dataset.isna().sum()
```

```python
weather_dataset = weather_dataset.interpolate(limit_direction='both')
weather_dataset.head()
```

```python
weather_dataset_np = weather_dataset.to_numpy()

for i in range(0, len(weather_dataset_np)):
    cloud = weather_dataset_np[i][3]
    
    if cloud >= 0 and cloud < 2:
        weather_dataset_np[i][3] = 0
    elif cloud >= 2 and cloud < 4:
        weather_dataset_np[i][3] = 2
    elif cloud >= 4 and cloud < 6:
        weather_dataset_np[i][3] = 4
    elif cloud >= 6 and cloud < 8:
        weather_dataset_np[i][3] = 6
    else:
        weather_dataset_np[i][3] = 8

weather_dataset_ready = pd.DataFrame(weather_dataset_np, columns=weather_dataset.columns)
weather_dataset_ready
```

```python
weather_dataset_ready['cloud'].unique()
```

```python
weather_valid = pd.read_csv('./../../../datasets/level3/weather_valid.csv')
weather_valid.head()
```

```python
weather_valid.dtypes
```

```python
weather_valid_test = weather_valid.merge(right=weather_dataset_ready, on='datetime', how='inner')
round(((((weather_valid_test['temperature_x'] - weather_valid_test['temperature_y']) ** 2).sum() / weather_valid_test.shape[0]) ** 0.5), 3)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_true=weather_valid_test.temperature_x, y_pred=weather_valid_test.temperature_y) ** 0.5
```
